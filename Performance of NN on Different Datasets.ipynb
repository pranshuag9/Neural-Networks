## Comparing Performance of Neural Network on Different Datasets

```python
from sklearn.datasets import make_moons,make_circles,make_classification
import neural_network as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_dataset(dataset):
    if dataset=='moons':
        X,Y = make_moons(n_samples=500,noise=0.2,random_state=1)
    elif dataset=='circles':
        X,Y = make_circles(n_samples=500,shuffle=True,noise=0.2,random_state=1,factor=0.2)
    elif dataset=='classification':
        X,Y = make_classification(n_samples=500,n_classes=2,n_features=2,n_informative=2,n_redundant=0,random_state=1)
    else:
        # Create XOR dataset
        X = np.array([
            [0,0],
            [0,1],
            [1,0],
            [1,1]
        ])
        Y = np.array([0,1,1,0])
        
    return X,Y
```

## Checking Performance with 2 Hidden Layers having 4 and 3 Neurons

```python
datasets = ["xor","classification","moons","circles"]

for d in datasets:
    model = nn.NeuralNetwork(input_size=2,layers=[4,3],output_size=2)
    X,Y = load_dataset(d)
    nn.train(X,Y,model,1000,0.001,logs=False)
    outputs = model.predict(X)
    
    train_acc = nn.training_accuracy(outputs,Y)
    
    plt.title("Dataset "+d)
    nn.plot_decision_boundary(lambda x:model.predict(x),X,Y)
    plt.show()
```

```python
for d in datasets:
    model = nn.NeuralNetwork(input_size=2,layers=[8,4],output_size=2)
    X,Y = load_dataset(d)
    nn.train(X,Y,model,1000,0.001,logs=False)
    outputs = model.predict(X)
    
    train_acc = nn.training_accuracy(outputs,Y)
    
    plt.title("Dataset "+d)
    nn.plot_decision_boundary(lambda x:model.predict(x),X,Y)
    plt.show()
```
