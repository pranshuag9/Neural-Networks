```python
import numpy as np
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
```

```python
input_size = 3 #No. of features
layers = [4,3]
output_size = 2
```

```python
def softmax(a):
    e_pa = np.exp(a) #e raise to the power a
    ans = e_pa/np.sum(e_pa,axis=1,keepdims=True)
    
    return ans
```

```python
a = np.array([[10,20]])

a_ = softmax(a)
print(a_)
```

```python
class NeuralNetwork:
    def __init__(self,input_size,layers,output_size):
        np.random.seed(0)
        
        model = {} #Dictionary
        
        #First Layer
        model['W1'] = np.random.randn(input_size,layers[0])
        model['b1'] = np.zeros((1,layers[0]))
        
        #Second Layer
        model['W2'] = np.random.randn(layers[0],layers[1])
        model['b2'] = np.zeros((1,layers[1]))
        
        #Third/Output Layer
        model['W3'] = np.random.randn(layers[1],output_size)
        model['b3'] = np.zeros((1,output_size))
        
        self.model = model
        
    def forward(self,x):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        b1,b2,b3 = self.model['b1'],self.model['b2'],self.model['b3']
        
        z1 = np.dot(x,W1) + b1
        a1 = np.tanh(z1) #Transforms input in range -1 to 1
        
        z2 = np.dot(a1,W2) + b2
        a2 = np.tanh(z2)
        
        z3 = np.dot(a2,W3) + b3
        y_ = softmax(z3) #Output
        
        #Store the outputs of each layer for back-propagation
        self.activation_outputs = (a1,a2,y_)
        
        return y_
        
    def backward(self,x,y,learning_rate=0.0001):
        # Creating Model Parameters
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        (a1,a2,y_) = self.activation_outputs
        
        # Computing error at final layer
        delta3 = y_ - y
        dw3 = np.dot(a2.T,delta3)
        db3 = np.sum(delta3,axis=0)
        
        # Computing the loss at 2nd hidden layer
        delta2 = (1-np.square(a2))*np.dot(delta3,W3.T)
        dw2 = np.dot(a1.T,delta2)
        db2 = np.sum(delta2,axis=0)
        
        # Computing the loss at 1st hidden layer
        delta1 = (1-np.square(a1))*np.dot(delta2,W2.T)
        dw1 = np.dot(x.T,delta1)
        db1 = np.sum(delta1,axis=0)       
        
        # Update the Model Parameters using Gradient Descent
        self.model["W1"] -= learning_rate*dw1
        self.model['b1'] -= learning_rate*db1
        
        self.model["W2"] -= learning_rate*dw2
        self.model["b2"] -= learning_rate*db2
        
        self.model["W3"] -= learning_rate*dw3
        self.model["b3"] -= learning_rate*db3
        
    def predict(self,x):
        y_out = self.forward(x)
        return np.argmax(y_out,axis=1)

    def summary(self):
        W1,W2,W3 = self.model['W1'],self.model['W2'],self.model['W3']
        a1,a2,y_ = self.activation_outputs
        
        
        print("W1 ",W1.shape)
        print("A1 ",a1.shape)
        print("W2 ",W2.shape)
        print("A2 ",a2.shape)
        print("W3 ",W3.shape)
        print("Y_ ",y_.shape)
```

```python
def loss(y_onehot,pred):
    l = -np.mean(y_onehot*np.log(pred))
    return l

def one_hot(y,depth):
    m = y.shape[0]
    y_onehot = np.zeros((m,depth))
    y_onehot[np.arange(m),y] = 1
    return y_onehot

def train(X,Y,model,epochs,learning_rate,logs=True):
    training_loss = []
    
    classes = len(np.unique(Y))
    y_oht = one_hot(Y,classes)
    
    for ix in range(epochs):
        Y_ = model.forward(X)
        l = loss(y_oht,Y_)
        training_loss.append(l)
        model.backward(X,y_oht,learning_rate)
        
        if(logs and ix%50==0):
            print("Epoch %d Loss %.4f"%(ix,l))
            
            
    return training_loss
```

```python
X,Y = make_circles(
    n_samples=500,
    shuffle=True,
    noise=0.05,
    random_state=1,
    factor=0.8
)
```

```python
plt.style.use("seaborn")
plt.scatter(X[:,0],X[:,1],c=Y,cmap=plt.cm.Accent)
plt.show()
```

```python
print("Shape of Y "+str(Y.shape))
y_oht = one_hot(Y,depth=3)
print(y_oht)
print("Shape of Y_One_Hot Notation "+str(y_oht.shape))
```

```python
model = NeuralNetwork(input_size=2,layers=[10,5],output_size=2)
```

```python
losses = train(X,Y,model,10000,0.0001)
```

```python
plt.plot(losses)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
model.summary()
```
